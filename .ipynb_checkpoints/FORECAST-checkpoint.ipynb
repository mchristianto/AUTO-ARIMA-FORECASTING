```python
import warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from pandas import datetime
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# PREPARING ALL DATA FRAME

df = pd.read_csv("train.csv", usecols=['date', 'item', 'sales'])
df = df.query('item > 0 and item < 6').reset_index(drop=True)
df.head()
```

```python
# SPLIT TO 5 DATA FRAME BASE ON ITEM_ID

df_1 = df.query('item == 1').reset_index(drop=True)
df_2 = df.query('item == 2').reset_index(drop=True)
df_3 = df.query('item == 3').reset_index(drop=True)
df_4 = df.query('item == 4').reset_index(drop=True)
df_5 = df.query('item == 5').reset_index(drop=True)
```

# SALES DATA 1

```python
#PREPROCESSING DATA

df_1['date'] = pd.to_datetime(df_1['date'])
df_1 = df_1.groupby(['date'])['sales'].sum().reset_index()
df_1 = df_1.set_index('date')
df_1 = df_1.resample('1M').sum()
df_1.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df_1)
plt.xlabel('Date', fontsize=13)
plt.ylabel('Sales', fontsize=13)
```

```python
#SPLIT DATA FOR TRAIN & TEST

df_1_train = df_1[:48]
df_1_test = df_1[47:60]
print("train = " + str(df_1_train.size) +" month")
print("test = " + str(df_1_test.size) +" month")
```

## AUTO ARIMA

```python
from pmdarima import auto_arima

model = auto_arima(df_1_train, start_p=1, start_q=1, m=12, seasonal=True,
                           d=1, D=1, trace=True,
                           error_action='ignore',  
                           suppress_warnings=True, 
                           stepwise=True)
print(model.aic())
```

```python
model.fit(df_1_train)
```

```python
prediction = model.predict(n_periods=13)
print(prediction)
```

```python
df_1_test.index
```

```python
df_prediction = pd.DataFrame(data=prediction)
df_prediction = df_prediction.set_index(df_1_test.index)
df_prediction.columns = ['sales']
df_prediction
```

```python
def MAPE(actual, prediction):
    actual, prediction = np.array(actual), np.array(prediction)
    return str("{0:0.4f}".format(np.mean(np.abs((actual - prediction) / actual)) * 100)) + "%"
```

```python
plt.figure(figsize=(8,5))
plt.plot(df_1_train, label='Train (actual)')
plt.plot(df_1_test, color='green', label='Test (actual)')
plt.plot(df_prediction, color='red', label='Prediction')
plt.legend(loc="upper left")
plt.title('SALES ITEM 1')
plt.tight_layout()

MAPE(df_1_test, df_prediction)
```
